#### Count the number of tweets by a given Twitter account.
Ask the user for the handle (@handle) of a twitter account. You will need to include a ***try/except block*** for account names not found. 
<br>***Hint:*** the program should count the number of tweets for any provided account.

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
````

````python
def tweets_function():
    url = 'https://twitter.com/'
    user = str(input('username: '))
    url = url + user
    response = requests.get(url)
    #get the try/except block for account names not found 
    try:
        response.status_code == 200
        html = response.content
        soup = BeautifulSoup(html)
        tweets_find = soup.find('li', {'class':'ProfileNav-item ProfileNav-item--tweets is-active'})
        tweets = int(tweets_find.find('a').find('span', {'class':'ProfileNav-value'})['data-count'])
        return tweets
    except:
        print('Username not found!')
        return False
````

#### Scrape a certain number of tweets of a given Twitter account.

````python
# This is the url you will scrape in this exercise 
# You will need to add the account credentials to this url
url = 'https://twitter.com/'
````

````python
def scrape_tweets():
    try:
        user = str(input("username:"))
        number_tweets = str(input("how many tweets should we scrape?"))
        
        #add the account credentials to this url
        response = requests.get(url + user)
        html = response.content
        soup = BeautifulSoup(html)
        tweets = soup.find_all('p', attrs = {'class':'TweetTextSize'})
        tweets = [tweet.text.replace('\xa0',' ') for tweet in tweets]
        return tweets[:int(number_tweets)]
    except:
        return "not found"
````

````python
scrape_tweets()
````
